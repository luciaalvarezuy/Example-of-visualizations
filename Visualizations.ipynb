## Bar Plots

Histogram of Age in order to explore the data of the selected variable

```python
intervals = range(min(df.Age), max(df.Age) + 2)
plt.figure(figsize=(12,8))
plt.hist(x=df.Age, bins=intervals, color='#F2AB6D', rwidth=0.85)
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(intervals)
plt.show() 
```

# Histogram and Density Plots

```python
sns.set(style="darkgrid")
sns.kdeplot(df['Age'], shade=True, bw=0.05, color='olive')
plt.show()
```

# Scatterplot

```python
sns.regplot(x=df["Education"], y=df["Age"])
```

# Heatmap

```python
plt.figure(figsize=(12,8))
correlations = df.corr(method='pearson')
sns.heatmap(correlations, cmap="coolwarm", annot=True)
plt.show()
```

# Box Plot 

```python
sns.boxplot( x=df["Gender"], y=df["Education"] )
```

```python
sns.set(style="darkgrid")
sns.boxplot(data=df.loc[:, ['Age', 'Education']])
plt.show()
```

# Violin Plot

```python
sns.violinplot(x=df["Gender"], y=df["Age"])
```
