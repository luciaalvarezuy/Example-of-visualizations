## Bar Plots

Histogram of Age in order to explore the data of the selected variable

```python
intervals = range(min(df.Age), max(df.Age) + 2)
plt.figure(figsize=(12,8))
plt.hist(x=df.Age, bins=intervals, color='#F2AB6D', rwidth=0.85)
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(intervals)
plt.show() 
```

## Histogram and Density Plots

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# assigning a graph to each ax
sns.boxplot(df["Age"], ax=ax_box)
sns.histplot(data=df, x="Age", ax=ax_hist)
 
# Remove x axis name for the boxplot
ax_box.set(xlabel='')
plt.show()
```

```python
sns.set(style="darkgrid")
sns.kdeplot(df['Age'], shade=True, bw=0.05, color='olive')
plt.show()
```
