Continúa con el conjunto de datos de tema deportivo que te guste y selecciona tres atributos del conjunto de datos. Calcula el p-valor y di si rechaza la hipótesis nula cogiendo un alfa de 5%.


Para realizar este ejercicio se opto por testear un ANOVA de una vía para datos independientes.
La variable a evaluar continua siendo Miles.
La variable de agrupación es Income. Para esto, se calcula los terciles de la variable Income.

```python
#Caclular los terciles de la variable Income
df['Terciles_de_Income'] = pd.qcut(df['Income'], q= 3, labels=False)

# Evaluar si se dividio correctamente 
df['Terciles_de_Income'].value_counts()
```

```python
df.iloc[45:60]
```

```python
# Explorar las variable Income según Terciles
df.groupby('Terciles_de_Income')['Miles'].agg(['mean', 'std', 'min', 'max'])
```

```python
# Representación de como se distribuyen los datos
fig, ax = plot.subplots(1, 1, figsize=(8, 4))
sns.boxplot(x="Terciles_de_Income", y="Miles", data=df, ax=ax)
sns.swarmplot(x="Terciles_de_Income", y="Miles", data=df, color='black', alpha = 0.5, ax=ax);
```

######  Primer supuesto a tester: Independencia
Presuponemos que la variable de agrupación (Income) y los deportistas dentro de cada grupo sin independiente entre ellos, ya que la muestra se obtuvo mediante un muestreo aleatorio.

######  Distribución normal de las observaciones

La variable cuantitativa debe de distribuirse de forma normal en cada uno de los grupos. El estudio de normalidad puede hacerse de forma gráfica (qqplot) o con test de hipótesis.

```python
# Gráficos qqplot
# ==============================================================================
fig, axs = plot.subplots(2, 2, figsize=(8, 7))
pg.qqplot(df.loc[df.Terciles_de_Income==0, 'Miles'], dist='norm', ax=axs[0,0])
axs[0,0].set_title('0')
pg.qqplot(df.loc[df.Terciles_de_Income==1, 'Miles'], dist='norm', ax=axs[0,1])
axs[0,1].set_title('1')
pg.qqplot(df.loc[df.Terciles_de_Income==2, 'Miles'], dist='norm', ax=axs[1,0])
axs[1,0].set_title('2')

plot.tight_layout()
```

```python
# Test de normalidad Shapiro-Wilk
pg.normality(data=df, dv='Miles', group='Terciles_de_Income')
```

```python
###### Tercer supuesto: Varianza constante entre grupos (homocedasticidad)
pg.homoscedasticity(data=df, dv='Miles', group='Terciles_de_Income', method='levene')
```

Observando los supuestos de normalidad, se considera que no se cumple el supuesto de normalidad y homoscedasticidad.
Por lo cual, se propone realizar la prueba de Kruskal-Wallis, que es un método no paramétrico.

Otra forma de continuar este trabajo, sería manipular la base de datos con el fin de extraer los casos extremos (outliers). Otro camino que se podría haber elegido podría ser la transformación de las variables.


###### Prueba de Kruskal-Wallis
A continuación, se realiza la prueba de  Kruskal-Wallis.

Las hipótesis a contrastar son:

H0= Las muestras provienen de poblaciones idénticas

H1= Las muestras provienen de poblaciones diferentes

```python
data = [df.loc[ids, 'Miles'].values for ids in df.groupby('Terciles_de_Income').groups.values()]
H, p = ss.kruskal(*data)
p
```

En este caso, el p valor es <.05, por lo cual, se podría rechazar la H0. Es decir, que se obtiene evidencia para concluir que existen diferencias en las millas recorridas por los deportistas según el income de los mismos. Para conocer en qué grupos hay diferencias significativas es necesario realizar un post hoc.


```python
#Post hoc
sp.posthoc_conover(df, val_col='Miles', group_col='Terciles_de_Income', p_adjust = 'holm')
```

```python
df.groupby('Terciles_de_Income')['Miles'].agg(['mean', 'std', 'min', 'max'])
```

En el posthoc se observa que hay diferencias estadísticamente significativas entre:
    - grupo 2 de income con grupo 0 y grupo 1

Esto quiere decir que las personas con mayor income (grupo 2 de Terciles_de_Income) reportan mayor promedio en Miles en comparación con el resto de los grupos, diferencia estadísticamente significativa.
